```python

# Fonction de traitement
def analyser_dates(df, nom):
    df = df.copy()

    # Conversion CPID en string
    df['CPID'] = df['CPID'].astype(str)

    # Conversion de Start en datetime
    df['Start'] = pd.to_datetime(df['Start'], errors='coerce')

    # Séparation en colonnes Date et Time
    df['Time'] = df['Start'].dt.strftime('%H:%M:%S')
    df['Start'] = df['Start'].dt.strftime('%Y-%m-%d')  

    # Colonnes optionnelles d'analyse
    df['Jour_semaine'] = pd.to_datetime(df['Start']).dt.day_name()
    df['Mois'] = pd.to_datetime(df['Start']).dt.month
    df['Heure (h)'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.hour

    # Statistiques affichées
    print(f"--- {nom} ---")
    print("Dates min :", df['Start'].min())
    print("Dates max :", df['Start'].max())
    print("Nb dates manquantes :", df['Start'].isna().sum())
    print("Répartition par mois :\n", df['Mois'].value_counts().sort_index())
    print()

    return df
```

```python
def analyser_dates(df, nom):
    df = df.copy()

    df['CPID'] = df['CPID'].astype(str)

    df['Start'] = pd.to_datetime(df['Start'], dayfirst=True, errors='coerce')

    def extraire_heure_avec_secondes(dt):
        if pd.isna(dt):
            return None
        # Extraire hh:mm puis ajouter :00 secondes
        return dt.strftime('%H:%M') + ':00'

    df['Time'] = df['Start'].apply(extraire_heure_avec_secondes)
    df['Start'] = df['Start'].dt.strftime('%Y-%m-%d')

    print(f"--- {nom} ---")
    print("Dates min :", df['Start'].min())
    print("Dates max :", df['Start'].max())
    print("Nb dates manquantes :", df['Start'].isna().sum())
    print()

    return df

```

```python
def analyser_dates_sep(df, nom):
    df = df.copy()

    df['CPID'] = df['CPID'].astype(str)
    df['Start'] = pd.to_datetime(df['Start'], errors='coerce')

    df['Date'] = df['Start'].dt.strftime('%Y-%m-%d')
    df['Time'] = df['Start'].dt.strftime('%H:%M:%S')

    # Supprimer la colonne Start d'origine
    df = df.drop(columns=['Start'])

    # Renommer Date en Start
    df = df.rename(columns={'Date': 'Start'})

    print(f"--- {nom} ---")
    print("Dates min :", df['Start'].min())
    print("Dates max :", df['Start'].max())
    print("Nb dates manquantes :", df['Start'].isna().sum())
    print()

    return df

```
